```python
#GREET PERSON
def greet(name):
    print(f"Hello,{name}! Welcome!")
greet("Ram")
```

```python
#FUNCTION TO ADD TWO NUMBERS
def add(a,b):
    return a+b
res=add(5,12)
print("Sum:",res)
```

```python
#FUNCTION TO CALCULATE AREA OF RECTANGLE
def calculate_area(l,b):
    return l*b
res=calculate_area(5,8)
print("Area of Rectangle:",res)
```

```python
#FUNCTION TO CHECK IF A NUMBER IS EVEN
def is_even(num):
    return num%2==0
print(is_even(6))
print(is_even(9))
```

```python
#FUNCTION TO GET MAXIMUM NUMBER FROM A LIST
def get_max(numbers):
    return max(numbers)
nums=[10,45,23,89,12]
print("Maximum value:",get_max(nums))
```
